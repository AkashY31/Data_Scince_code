```python
#imported libraries
import numpy as np  
import pandas as pd  
import matplotlib.pyplot as plt  
%matplotlib inline
```

```python
#imported online dataset
from sklearn.datasets import load_breast_cancer
df = load_breast_cancer()
df
```

```python
#splitting data
x= df.data
y = df.target
```

```python
#splling data in x and y part
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.30,random_state=42)
```

```python
#imported decision Tree classifier
from sklearn.tree import DecisionTreeClassifier
treeClasifier = DecisionTreeClassifier()
```

```python
#fitting data in x and y part  
treeClasifier.fit(x_train,y_train)
```

```python
from sklearn import  tree
```

```python
plt.figure(figsize=(10,7))
tree.plot_tree(treeClasifier,filled=True)
```

```python
y_pred = treeClasifier.predict(x_test)
y_pred
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_pred,y_test).round(2)
```

#Post-proning 

```python
#Post-proning with max_depth = 2
treeClasifier = DecisionTreeClassifier(max_depth=2)
```

```python
#fitting data 
treeClasifier.fit(x_train,y_train)
```

```python
plt.figure(figsize=(10,8))
tree.plot_tree(treeClasifier,filled=True)
```

```python
y_pred = treeClasifier.predict(x_test)
y_pred
```

```python
accuracy_score(y_pred,y_test).round(2)
```

#Using Entropy

```python
from sklearn.tree import DecisionTreeClassifier
treeClasifier = DecisionTreeClassifier(criterion="entropy",max_depth=2)
```

```python
#data-fitting
treeClasifier.fit(x_test,y_test)
```

```python
#plotting tree
plt.figure(figsize=(12,9))
tree.plot_tree(treeClasifier,filled=True)
```

```python
y_pred = treeClasifier.predict(x_test)
y_pred
```

```python
accuracy_score(y_pred,y_test)
```
