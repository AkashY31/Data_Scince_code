```python
#importing required library
import pandas as pd   
import numpy as np  
import matplotlib.pyplot as plt  
%matplotlib inline
```

```python
#importing datasets from sklearn
from sklearn.datasets import load_iris
iris_df = load_iris()
print(iris_df.DESCR)
```

```python
#importing seaborn library
import seaborn as sns 
```

```python
#importing datasets using seaborn library
df = sns.load_dataset('iris')
df.head()
```

```python
#splitting- datasets
x=df.iloc[:,0:-1]
y = df.iloc[:,-1]
```

```python
#spliiting datasets using train-test-split
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.33,random_state=42)
```

```python
from sklearn.tree import DecisionTreeClassifier
des_tree_clssfr = DecisionTreeClassifier()
```

```python
des_tree_clssfr.fit(x_train,y_train)
```

```python
from sklearn import tree 
plt.figure(figsize=(15,10))
tree.plot_tree(des_tree_clssfr,filled=True)
```

```python
#Post-Proning with max_depth 2
from sklearn.tree import DecisionTreeClassifier
treeClassifier = DecisionTreeClassifier(max_depth=2)
```

```python
treeClassifier.fit(x_train,y_train)
```

```python
plt.figure(figsize=(15,10))
tree.plot_tree(treeClassifier,filled=True)
```

```python
#prediction 
y_pred = treeClassifier.predict(x_test)
y_pred
```

```python
#verifying result
from sklearn.metrics import accuracy_score,classification_report
```

```python
print(accuracy_score(y_test,y_pred)) #accuracy
```

```python
print(classification_report(y_test,y_pred)) #classification_Report
```

#construction of Tree using Entropy

```python
tree_entropy_clsfr = DecisionTreeClassifier(criterion="entropy", max_depth=2)
```

```python
tree_entropy_clsfr.fit(x_train,y_train)
```

```python
plt.figure(figsize=(15,10))
tree.plot_tree(tree_entropy_clsfr,filled=True)
```

## Decision Preproning and Hyperparameter Tuning For Huge-data

```python
import warnings
warnings.filterwarnings('ignore')
```

```python
#defined parameters
parameter={
    'criterion':['gini','entropy','log_loss'],
    'splitter':['best','random'],
    'max_depth':[1,2,3,4,5],
    'max_features':['auto','sqrt','log2']
}
```

```python
#using - GridSearchCV
from sklearn.model_selection import GridSearchCV
```

```python
tree_classifier = DecisionTreeClassifier()
gsc = GridSearchCV(tree_classifier,param_grid=parameter,cv=5,scoring='accuracy')
```

```python
#Training the dataset
gsc.fit(x_train,y_train)
```

```python
gsc.best_params_
```

```python
y_pred = gsc.predict(x_test)
y_pred
```

```python
accuracy_score(y_pred,y_test)
```

```python
classification_report(y_pred,y_test)
```
