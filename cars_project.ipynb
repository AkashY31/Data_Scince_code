```python
import numpy as np 
import pandas as pd 
import seaborn as sns 
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("Toyoto_Corrola.csv",index_col=0)
data.describe()
```

```python
data.columns
```

```python
x = data[['Age_08_04', 'KM', 'HP', 'Doors', 'Cylinders',
       'Gears', 'Weight']]
```

```python
x.head()
```

```python
y = data[['Price']]
y.head()
```

```python
import sklearn as sk  
from sklearn.model_selection import  train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2)
```

```python
print(x_train.shape)
print(x_test.shape)
print()
print(y_train.shape)
print(y_test.shape)
```

```python
from sklearn.linear_model import LinearRegression
lm=LinearRegression()
```

```python
lm.fit(x_train,y_train)
```

```python
print("cofiecient is",  lm.coef_)
print()
print("intercept is",  lm.intercept_)
```

```python
y_pred = lm.predict(x_test)
y_pred
```

```python
new_data=pd.DataFrame()
new_data= x_test
new_data['Price'] = y_test
new_data['Actual Price'] = y_pred
new_data
```

```python
lm.score(x_train,y_train)*100
```
